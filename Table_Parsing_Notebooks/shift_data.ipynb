```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('/Volumes/Extreme SSD/Python_Projects/NIAA Project/data/data_original_structure.xlsx')
```

```python
df.head()
```

```python
# Shift the data to the left by one position for the last 3 columns starting from row 157
#df.loc[157:, df.columns[-4:]] = df.loc[157:, df.columns[-3:]].shift(-1, axis=1)
row_start = 157

# Step 1: Identify the last 4 columns
rightmost_four = df.columns[-4:]  # e.g., [A, B, C, D]

# Step 2: Shift those 4 columns left
temp_shifted = df.loc[row_start:, rightmost_four].shift(-1, axis=1)

# Step 3: Assign shifted data back to the same columns
df.loc[row_start:, rightmost_four] = temp_shifted
```

```python
# Drop the last column
df = df.drop(df.columns[-1], axis=1)
```

```python
# Write the data to a new excel file
df.to_csv('/Volumes/Extreme SSD/Python_Projects/NIAA Project/data/data_unstructured.csv', index=False)
```
