## Split the columns into groups
Every 7 columns is considered a group. (This is because the original data was strcutured in a way where it had 7 columns). 


```python
# Determine the index where you want to split the DataFrame
split_index = 7

# Split the DataFrame into two
df1 = df.iloc[:, :split_index]
df2 = df.iloc[:, split_index:]
```

```python
df1.head()
```

```python
df2.head()
```

```python
df2.replace(['', 'NaN', 'None', None], np.nan, inplace=True)
```

```python
df2.dropna(how='all', inplace=True)
```

```python
df2.head()
```

```python
# Write df2 to an Excel file
df2.to_excel('/Volumes/Extreme SSD/Python_Projects/NIAA Project/ALLE_ZENDINGSARBEIDERS_CLEANED_data_3.xlsx', index=True)
```

```python
df2.head()
```

```python

# Assuming df is your original DataFrame with 42 columns

# Initialize lists to collect rows for the new DataFrames
spouses = []  # Subgroups where the first column is "~"
events = []  # All other subgroups

# Iterate over each row in the original DataFrame
for idx, row in df2.iterrows():
    
    # Split the row into 6 subgroups of 7 columns each
    for group_num in range(6):
        start_col = group_num * 7
        end_col = start_col + 7
        subgroup = row.iloc[start_col:end_col].tolist()
        first_value = subgroup[0]
        
        # Prepare the new row with original index and subgroup data
        new_row = [idx] + subgroup
        print(first_value)
        # Check if the first column of the subgroup contains "~"
        if isinstance(first_value, str) and "~" in first_value:
            spouses.append(new_row)
        else:
            events.append(new_row)

# Define columns for the new DataFrames
columns = ['original_index'] + [f'Col{i+1}' for i in range(7)]

# Create the new DataFrames
spouses_df = pd.DataFrame(spouses, columns=columns)
events_df = pd.DataFrame(events, columns=columns)

# Display the results
print("DataFrame 1 (Subgroups with '~'):")
spouses_df.head()
  
```

```python
# Write the DataFrames to exel files

spouses_df.to_excel('spouses.xlsx', index=False)
events_df.to_excel('events.xlsx', index=False)

```
